```python
class Automovil:
    def __init__(self, marca, placas, precio):
        self.marca = marca
        self.placas = placas
        self.precio = precio

    def leerdatos(self):
      self.marca = input("Ingrese la marca: ")
      self.placas = input("Ingrese las placas: ")
      self.precio = input("Ingrese el precio: ")

    def mostrardatos(self):
      print("Marca:", self.marca)
      print("Placas:", self.placas)
      print("Precio:", self.precio)

# fin de la clase automovil

# creando los objetos carro1, carro2, carro3

print("Carro Numero 1")
carro1 = Automovil("", "", 0.0) # creando el objeto carro1 con valores iniciales
carro1.leerdatos()
carro1.mostrardatos()
print("-+-+-+-+-+-+-+-+--+-+-+-+---+-+-+-+-+-+-")

print("Carro Numero 2")
carro2 = Automovil("", "", 0.0) # creando el objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print("-+-+-+-+-+-+-+-+--+-+-+-+---+-+-+-+-+-+-")

print("Carro Numero 3")
carro3 = Automovil("", "", 0.0) # creando el objeto carro3
carro3.leerdatos()
carro3.mostrardatos()
print("-+-+-+-+-+-+-+-+--+-+-+-+---+-+-+-+-+-+-")
print("Software realizado por Iris Chavez NL = 11")
```
